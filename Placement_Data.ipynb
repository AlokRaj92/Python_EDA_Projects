# EDA :-

```python
ax = sns.countplot(data = df, x = 'hsc_s',hue = 'hsc_s')
for i in ax.containers:
    ax.bar_label(i)
plt.title("Subject by Students",color = 'red', size = 25)
plt.xlabel("Subjects", color = "red", size =20)
plt.ylabel("Student Counts", color = "red", size =20)
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'degree_t',hue = 'degree_t')
for i in ax.containers:
    ax.bar_label(i)
plt.title("Degree by Students",color = 'red', size = 25)
plt.xlabel("Subjects", color = "red", size =20)
plt.ylabel("Student Counts", color = "red", size =20)
plt.show()
```

```python
ax = sns.countplot(data = df,x = 'gender',hue = 'gender')
for i in ax.containers:
    ax.bar_label(i)
plt.title("Gender by Students",color = 'red', size = 25)
plt.xlabel("Gender", color = "red", size =20)
plt.ylabel("Gender Counts", color = "red", size =20)
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'specialisation',hue = 'specialisation')
for i in ax.containers:
    ax.bar_label(i)
plt.title("Specialisation by Students",color = 'red', size = 25)
plt.xlabel("Specialisation", color = "red", size =20)
plt.ylabel("Student Counts", color = "red", size =20)
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'status', hue = 'status')
for i in ax.containers:
    ax.bar_label(i)
plt.title("Selection of Students",color = 'red', size = 25)
plt.xlabel("Students", color = "red", size =20)
plt.ylabel("Student Counts", color = "red", size =20)
plt.show()
```

```python
sns.set(rc = {'figure.figsize':(5,4)})
ax = sns.countplot(df[df['status']== 'Not Placed'],x = 'gender',hue = 'gender')
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("UnPlaced Students",color = 'purple', size = 20)
plt.xlabel("Gender",color = 'purple', size = 20)
plt.ylabel("Count",color = 'purple', size = 20)
plt.grid(False)
plt.show()
```

# Salary by Specialisation

```python
salary_by_speci = df.groupby(['specialisation'],as_index = False)['salary'].max().sort_values(by = 'salary', ascending = False)
sns.set(rc = {'figure.figsize':(5,4)})
ax = sns.barplot(data =salary_by_speci, x = 'specialisation', y = 'salary',hue = 'specialisation',palette = 'bright')
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

# According to above chart we undaerstand that "Marketing & Finance" department gets high LPA.

# Degree and MBA percentage based Salary

```python
salary_degree_mba = df.groupby(['degree_t','mba_p'],as_index = False)['salary'].max().sort_values(by = 'salary', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,6)})
ax = sns.barplot(data = salary_degree_mba, x ='salary', y = 'mba_p', hue = 'degree_t',palette = 'bright')
for bar in ax.containers:
    ax.bar_label(bar)
plt.legend(loc = 'best', bbox_to_anchor = (1,1))
plt.show()
```

# In above chart we can say that "Commerce & Managment" student who got more than "70 %" are placed at only 4LPA.

```python
salary_by_gender = df.groupby(['gender'],as_index = False)['salary'].max().sort_values(by = 'salary', ascending = False)
sns.set(rc = {"figure.figsize":(5,4)})
ax = sns.barplot(data = salary_by_gender, x = 'gender', y = 'salary',hue = 'gender',palette ='Set1')
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Gender by Salary", size = 20, color = 'green')
plt.xlabel("Gender",color = "green",size = 20)
plt.ylabel("Salary", color = "green", size = 20)
plt.grid(False)
plt.show()
```

# In above chart "Male" candidate gets high LPA.

```python
gender_by_status = df.groupby(['gender'],as_index = False)['status'].count().sort_values(by = 'gender', ascending = False)
x = sns.barplot(data = gender_by_status, x = 'gender', y = 'status', hue = 'gender', palette = 'Set2')
for i in x.containers:
    x.bar_label(i)
plt.show()
```

# According to above chart the number of placement of "Male" Students are higher than "Female"

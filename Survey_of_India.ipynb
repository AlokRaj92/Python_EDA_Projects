# Top 5 higest & lowest Populated State:-

```python
top_5 = df.groupby(['state_name'],as_index = False)['population_total'].sum().sort_values(by = 'population_total', ascending = False).head()
sns.set(rc = {'figure.figsize':(10,5)})
x = sns.barplot(data = top_5, x = 'state_name', y = 'population_total', hue = 'population_total',palette = 'rainbow', legend = False)
for bar in x.containers:
    x.bar_label(bar)
plt.title("Top 5 higest populated States", color = "red", size = 20)
plt.xlabel("States Name", color = "red", size = 15)
plt.ylabel("Population", color = "red", size = 15)
plt.grid(False)
plt.show()
```

```python
tail_5 = df.groupby(['state_name'],as_index = False)['population_total'].sum().sort_values(by = 'population_total', ascending = False).tail()
sns.set(rc = {'figure.figsize':(10,5)})
x = sns.barplot(data = top_5, x = 'state_name', y = 'population_total', hue = 'population_total',palette = 'bright', legend = False)
for bar in x.containers:
    x.bar_label(bar)
plt.title("Top 5 lowest populated States", color = "red", size = 20)
plt.xlabel("States Name", color = "red", size = 15)
plt.ylabel("Population", color = "red", size = 15)
plt.xticks(rotation = 15)
plt.grid(False)
plt.show()
```

# In above both graph we find that "Maharashtra" has higest population and "Andaman & Nicobar" has lowest population.
